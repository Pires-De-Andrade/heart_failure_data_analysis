# 02 — Análise de Variáveis Qualitativas

**Variáveis:** `Sex`, `ChestPainType`, `RestingECG`, `ExerciseAngina`,
`ST_Slope`, `FastingBS`

Para cada variável:
1. Tabela de frequência absoluta e relativa
2. Gráfico de barras (distribuição geral)
3. Tabela cruzada com HeartDisease
4. Gráfico de barras agrupadas (cruzamento com HeartDisease)
5. Comentário analítico

---

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
import matplotlib.pyplot as plt

from src.config import QUAL_VARS, TARGET, VAR_LABELS, CAT_LABELS, OUTPUT_TABLES
from src.data_loader import load_subsets
from src.frequency_tables import freq_table_qualitative, freq_table_crosstab
from src.plotting import bar_chart, bar_chart_grouped

data = load_subsets()
df = data["df_full"]
print(f"Dataset carregado: {df.shape[0]} observações, {df.shape[1]} variáveis")
```

---
## 2.1 Sexo (`Sex`)

```python
var = "Sex"
print(f"\n{'='*60}")
print(f"  {VAR_LABELS[var]}")
print(f"{'='*60}\n")

# Tabela de frequência
ft = freq_table_qualitative(df[var])
display(ft)
ft.to_csv(OUTPUT_TABLES / f"freq_{var}.csv")
```

```python
# Gráfico de barras
bar_chart(df[var], var, filename=f"02_bar_{var}.png")
plt.show()
```

```python
# Tabela cruzada com HeartDisease
ct = freq_table_crosstab(df, var)
display(ct)
ct.to_csv(OUTPUT_TABLES / f"crosstab_{var}.csv")
```

```python
# Barras agrupadas
bar_chart_grouped(df, var, filename=f"02_grouped_{var}.png")
plt.show()
```
