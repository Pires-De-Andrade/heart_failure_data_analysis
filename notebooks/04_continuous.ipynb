# 04 — Análise de Variáveis Quantitativas Contínuas

**Variáveis:** `RestingBP`, `Cholesterol`, `MaxHR`, `Oldpeak`

Para cada variável:
1. Classes pela regra de Sturges (k = 1 + 3,322 × log₁₀n)
2. Tabela de frequências (fi, fri, Fi, Fri)
3. Histograma com curva de densidade
4. Medidas descritivas completas (dados agrupados em classes)
5. Boxplot comparativo por HeartDisease

**Coeficiente de assimetria:** Pearson 2 → AS = 3 × (média − mediana) / σ

### Subsets utilizados
- `RestingBP`: n = 917 (excluída 1 linha com valor 0)
- `Cholesterol`: n = 746 (excluídas 172 linhas com valor 0)
- `MaxHR`, `Oldpeak`: n = 918 (dataset completo)

---

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as sp_stats

from src.config import CONT_VARS, TARGET, VAR_LABELS, OUTPUT_TABLES
from src.data_loader import load_subsets
from src.frequency_tables import freq_table_classes, sturges_k, sturges_bins
from src.descriptive_stats import (
    descriptive_summary, descriptive_by_group,
    pearson2_skewness, interpret_skewness, interpret_kurtosis,
)
from src.plotting import histogram, boxplot_comparison

data = load_subsets()
df_full = data["df_full"]
df_chol = data["df_chol"]
df_bp   = data["df_bp"]

# Mapeamento variável → dataset correto
var_datasets = {
    "RestingBP":   df_bp,
    "Cholesterol": df_chol,
    "MaxHR":       df_full,
    "Oldpeak":     df_full,
}

for var, ds in var_datasets.items():
    print(f"{var}: n = {len(ds)}")
```

---
## 4.1 Pressão Arterial em Repouso (`RestingBP`)

n = 917 (excluída 1 observação com RestingBP = 0)

```python
var = "RestingBP"
ds = var_datasets[var]
series = ds[var]
n = len(series)
k = sturges_k(n)

print(f"{'='*60}")
print(f"  {VAR_LABELS[var]}")
print(f"{'='*60}")
print(f"n = {n} | k (Sturges) = {k}")
print(f"Min = {series.min()} | Max = {series.max()}")
print(f"Amplitude = {series.max() - series.min()}")
```

```python
# Tabela de frequências com classes de Sturges
ft = freq_table_classes(series)
display(ft)

assert ft["fi"].sum() == n, f"Erro: soma de fi ({ft['fi'].sum()}) ≠ n ({n})"
print(f"\n✓ Soma de fi = {ft['fi'].sum()}")

ft.to_csv(OUTPUT_TABLES / f"freq_{var}_classes.csv")
```

```python
# Histograma com curva de densidade
bins = sturges_bins(series)
histogram(series, var, bins=bins, density_curve=True,
          filename=f"04_hist_{var}.png")
plt.show()
```

```python
# Medidas descritivas
desc = descriptive_summary(series, name=VAR_LABELS[var])
display(desc)
desc.to_csv(OUTPUT_TABLES / f"desc_{var}.csv")

# Interpretação
as_val = pearson2_skewness(series)
kurt_val = sp_stats.kurtosis(series, bias=False)
print(f"\nAssimetria (Pearson 2): {as_val:.4f} → {interpret_skewness(as_val)}")
print(f"Curtose (excesso): {kurt_val:.4f} → {interpret_kurtosis(kurt_val)}")

# Verificação interna — scipy.stats.skew (NÃO reportado)
# skew_scipy = sp_stats.skew(series, bias=False)
# print(f"[Verificação] scipy.stats.skew: {skew_scipy:.4f}")
```

```python
# Boxplot comparativo por HeartDisease
boxplot_comparison(ds, var, filename=f"04_boxplot_{var}_vs_HD.png")
plt.show()

# Descritivas por grupo
desc_grp = descriptive_by_group(ds, var)
display(desc_grp)
desc_grp.to_csv(OUTPUT_TABLES / f"desc_{var}_por_grupo.csv")
```
