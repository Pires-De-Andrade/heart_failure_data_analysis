# 03 — Análise de Variável Quantitativa Discreta

**Variável:** Idade (`Age`)

Neste dataset, a idade é tratada como quantitativa discreta (valores
inteiros em anos).

**Abordagem metodológica exigida:**
1. **Tabelas e Gráficos:** Agrupamento em classes (faixas etárias) para
   facilitar a visualização (Regra de Sturges não se aplica bem a idades
   na clínica; usaremos cortes decenais clássicos).
2. **Medidas Descritivas:** Calculadas rigorosamente sobre os valores
   **brutos** (n=918), não pelas fórmulas de dados agrupados.
3. **Coeficiente de Assimetria:** O foco será no **Segundo Coeficiente
   de Pearson (Pearson 2)** calculado como: `AS = 3 * (média - mediana) / σ`.

---

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as sp_stats

from src.config import (
    DISC_VAR, TARGET, VAR_LABELS, AGE_BINS, AGE_LABELS, OUTPUT_TABLES
)
from src.data_loader import load_subsets
from src.frequency_tables import freq_table_classes
from src.descriptive_stats import (
    descriptive_summary, descriptive_by_group, 
    pearson2_skewness, interpret_skewness, interpret_kurtosis
)
from src.plotting import histogram, boxplot_comparison

data = load_subsets()
df = data["df_full"]
var = DISC_VAR
series = df[var]
```

## 3.1 Tabela de frequências (Faixas Etárias)

Categorização manual: <40, 40-49, 50-59, 60-69, ≥70

```python
ft_age = freq_table_classes(series, bins=AGE_BINS, labels=AGE_LABELS)
display(ft_age)
ft_age.to_csv(OUTPUT_TABLES / "freq_Age_classes.csv")
```

## 3.2 Visualização da Distribuição (Histograma)

```python
histogram(series, var, bins=15, density_curve=True, 
          filename="03_hist_Age.png")
plt.show()
```

## 3.3 Medidas Descritivas (Valores Brutos)

Cálculo exato sobre as 918 observações individuais.

```python
desc = descriptive_summary(series, name=VAR_LABELS[var])
display(desc)
desc.to_csv(OUTPUT_TABLES / "desc_Age_bruto.csv")
```

```python
# Cálculo da Assimetria de Pearson 2
as_pearson2 = pearson2_skewness(series)
kurt = sp_stats.kurtosis(series, bias=False)

print(f"Assimetria (Pearson 2): {as_pearson2:.4f}")
print(f"Interpretação: {interpret_skewness(as_pearson2)}\n")

print(f"Curtose (excesso): {kurt:.4f}")
print(f"Interpretação: {interpret_kurtosis(kurt)}\n")

# Verificação interna com scipy.stats.skew (NÃO reportar como valor principal)
# print(f"[Verificação] scipy.stats.skew: {sp_stats.skew(series, bias=False):.4f}")
```

## 3.4 Comparativo: Idade × Doença Cardíaca

```python
# Boxplot comparativo
boxplot_comparison(df, var, filename="03_boxplot_Age_vs_HD.png")
plt.show()
```

```python
# Tabela de descritivas dividida por grupo (com doença vs sem doença)
desc_grp = descriptive_by_group(df, var)
display(desc_grp)
desc_grp.to_csv(OUTPUT_TABLES / "desc_Age_por_grupo.csv")
```
